# PART 8 - Grouping and Aggregating - Analyzing and Exploring Your Data

```python
import pandas as pd
df = pd.read_csv('survey_results_public.csv', index_col='Respondent')
schema_df = pd.read_csv('survey_results_schema.csv', index_col='Column')
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
```

```python
df[df['Country']=='Egypt']['Country'].value_counts() #filter rows to include only egypt and then find # of egyptians in the survey
```

```python
country_grp = df.groupby('Country')
```

```python
country_grp.get_group('Egypt')['Country'].value_counts() #using groupby
```

```python
filt = df['Country'] == 'Egypt'
df.loc[filt]['SocialMedia'].value_counts()
```

```python
##use loc with groupby after applying a function since the index becomes the column we are grouping by
country_grp['SocialMedia'].value_counts().loc['Egypt'] 
```

```python
#seriesGroupby (each country has a ConvertedComp series of its own --> list of series)
country_grp['ConvertedComp']
```

```python
country_grp['ConvertedComp'].median().sort_values(ascending=False) # find largest salary medians per country
```

```python
#Egypt median salary
country_grp['ConvertedComp'].median().loc['Egypt']
```

```python
#applying multiple aggregation on the same group using agg()
country_grp['ConvertedComp'].agg(['median', 'mean'])
```

```python
country_grp['ConvertedComp'].agg(['median', 'mean']).loc['Egypt']
```

```python
filt = df['Country'] == 'Egypt'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum() # no. of egyptians who worked with python
```

```python
# error --> that's a series group by so you need to loop in each series and perform sum on each
country_grp['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
#use apply with lambda function
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
#for each series x in the group by series find the sum of developers who worked with python
```

```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()).loc['Egypt']
```

```python
#find the ratio between developers who worked with python to the total developers per country
country_respondents = df['Country'].value_counts()
print(country_respondents)
```

```python
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
print(country_uses_python)
```

```python
#merge two series on two separate columns (axis = 'columns')
python_df = pd.concat([country_respondents , country_uses_python], axis = 'columns', sort=False)
python_df
```

```python
python_df.rename(columns={'count' : 'NumRespondents', 'LanguageWorkedWith' : 'NumKnowsPython'}, inplace = True )
```

```python
python_df
```

```python
#add a ratio column
python_df['PythonRespondentRatio'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100
```

```python
python_df
```

```python
python_df.sort_values(by = 'PythonRespondentRatio', ascending= False , inplace= True)
```
